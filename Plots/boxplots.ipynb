```python
import statistics
import scipy.stats as stats
import matplotlib.pyplot as plt
import os

#To plot the right data, just follow the instructions in fitness-plots.ipynb

def yLast(file, n_lev):
    file = open(file, "r")
    a = file.read()
    file.close()
    a = a.split(" ") #Ancora array di stringhe
    offset = n_lev*1000
    i = 99
    y = []
    for j in range(20):
        y.append(float(a[offset+i+j*100])) #Offset per il bug di scrittura nei file
    return y

n = "0"
n2 = "0" #Output file number
i_best = 1
files = ["qmo_"+n+"_6.txt", "unif_"+n+".txt", "1p_"+n+".txt", "2p_"+n+".txt"]
indexes = [i_best, 0, 0, 0]
x = []
for i in range(len(files)):
    x.append(yLast(files[i], indexes[i]))
#plt.figure(dpi=300)
plt.boxplot(x, labels=["QMO (best)","UC","OPC","TPC"])
plt.ylabel("Fitness")
plt.title("Solution fitness vs operator - Instance #"+n2)
plt.savefig(os.getcwd()+"\Boxplots\Bplot-"+n2+".pdf")
plt.show()
```

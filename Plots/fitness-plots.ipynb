To re-create the plots shown in the paper, manually choose the best noise level for QMO (plotted with a continuous line) and choose the number of the instance (**n** variable-> note that it's a string). The noise level **i_best** ranges from 0 to 6, for $p_{bf}=0, ..., 0.3$.

```python
import matplotlib.pyplot as plt
plt.figure(dpi=300)
x = range(100)
plt.title("Mean fitness vs generation plot - Instance #"+n2)
colours = ['dark blue', 'orange', 'green', 'red', 'purple', 'brown', 'fuchsia', 'gray', 'goldenrod', 'sky blue']
i_best = 1 #Da variare per ogni grafico (vedi a mano miglior noise level)
for i in range(7): #Tenere fuori il livello a performance migliore (nel loop linee tratteggiate)
    if i!=i_best:
        plt.plot(x, f[i], label=("QMO ($p_{bf}="+str(i/20)+"$)"), linestyle="dashed", alpha=0.5, color = ('xkcd:'+colours[i]))
#Aggiungere gli altri operatori
plt.plot(x, f[i_best], label=("QMO ($p_{bf}="+str(i_best/20)+"$)"), color = ('xkcd:'+colours[i_best]))
plt.plot(x, f[7], label="UC", color = ('xkcd:'+colours[7]))
plt.plot(x, f[8], label="OPC", color = ('xkcd:'+colours[8]))
plt.plot(x, f[9], label="TPC", color = ('xkcd:'+colours[9]))
plt.ylabel("Mean fitness")
plt.xlabel("Generation")
plt.legend(fontsize="x-small")
plt.savefig("plot-"+n2+".pdf")
plt.show()
```
